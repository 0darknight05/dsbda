```python
# 1
import pandas as pd
df = pd.read_csv("./student_sample_data.csv")
df.head()

# 2
print(df.isnull().sum())
print(df.describe())
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])

# 3
print(df.dtypes)
df['Age']=df['Age'].astype('category')
df['Ethnicity']=df['Ethnicity'].astype('category')
df['GPA']=df['GPA'].astype('category')
df['Gender']=df['Gender'].astype('category')
print(df.dtypes)

# 4
df_encoded=pd.get_dummies(df,columns=['Age','GPA','Gender'])
df_encoded.head()
```
