```python
import pandas as pd
import numpy as np
df = pd.read_csv("./student_sample_data.csv")
df.head()


min_age = df['Age'].min()
print("Minmum Age:", min_age)

max_age = df['Age'].max()
print("Max Age:", max_age)

mean_age = df['Age'].mean()
print("Mean Age:", mean_age)

median_age = df['Age'].median()
print("Median Age:", median_age)

sd_age = df['Age'].std()
print("Standard deviation of Age:", sd_age)

percentage = (df['Age'] == 18).sum() / len(df) * 100
print(percentage)
```
