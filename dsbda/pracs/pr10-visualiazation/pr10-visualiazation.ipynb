```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------
# Load dataset (inbuilt from seaborn)
# -------------------------------
df = sns.load_dataset('iris')

print(df.head())

# -------------------------------
# 1. Features and their types
# -------------------------------
print("\nFeature Types:\n")
print(df.dtypes)

# -------------------------------
# 2. Histogram for each feature
# -------------------------------
df.hist(bins=20, edgecolor='black', figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()

# -------------------------------
# 3. Boxplot for each feature
# -------------------------------
for column in df.columns[:-1]:  # exclude 'species'
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()

# -------------------------------
# 4. Compare distributions
# -------------------------------
sns.pairplot(df, hue='species')
plt.show()
```
